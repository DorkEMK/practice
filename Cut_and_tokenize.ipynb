### Choose chapters

```python
import os
import pandas as pd
import gc
import csv
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
punct = '!"#$%&()*+,-./:;<=>?@[\]^_`{|}~'
```

```python
def tokenize(text):
    
    words = [word.strip(punct) for word in text.split() if word != '']
    words = [word for word in words if word]

    return words
```

```python
searchfor = ['CHAPTER ', 'PART ', 'Chapter ', 'Part ', 'BOOK ']
```

```python
for file in os.listdir():
    print(file)
    if file.endswith(".tsv"):
        df = pd.read_csv(file, sep = '\t', encoding = 'UTF8', error_bad_lines=False, index_col=0)
        
        #строки, подозрительные на начало главы
        print(df[df['en'].str.contains('|'.join(searchfor))])
        
        # по какую главу обрезать
        num = int(input())
        
        if num == 0:
            df_cut = df.copy()
        else:
            df_cut = df[0:num-1].copy()
        # токенизируем
        df_cut['ru_tokens'] = df_cut['ru'].apply(tokenize)
        
        filename = str(file)[:-4]+'_cut.tsv'
        df_cut.to_csv(filename, sep = '\t')
        
        # не оставляем df в памяти
        del df, df_cut
        gc.collect()
```
