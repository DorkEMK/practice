```python
import os, sys, pandas as pd, json, gc
```

```python
from ast import literal_eval
for file in os.listdir():

    if file.endswith(".tsv"):
        print(file)
        data = []
        df = pd.read_csv(file, sep = '\t', encoding = 'UTF8', error_bad_lines=False, index_col=0)

        for index, row in df.iterrows():
            obj={}
            obj['en']=row['en']
            obj['ru']=row['ru']
            obj['ru_tokens']=literal_eval(row['ru_tokens'])
            #print(row['en_tokens'])
            obj['en_tokens']=literal_eval(row['en_tokens'])
            
            
            data.append(obj)
    
        with open(file[:-4] + '.json', 'w') as outfile:  
            json.dump(data, outfile)
                
        del df
        gc.collect()        
```
